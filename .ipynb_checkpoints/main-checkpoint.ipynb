```python
from pandas import DataFrame
import pandas as pd
import numpy as np
```

```python
data = pd.read_table('./docs/export_low_small.txt')
```

```python
# 展示前 10 行数据
data[:10]
```

```python
# 连接 mongodb
from pymongo import MongoClient
conn = MongoClient('localhost', 27017)
db = conn.mydb
my_set = db.test_set
```

```python
# 向mongodb插入数据
tobeInser = data.values[1:,:]
totalNum = tobeInser.shape[0]
for i in range(totalNum):
    my_set.insert({"freq":int(tobeInser[i,0]),
                   "phi":float(tobeInser[i,1]),
                   "vals":tobeInser[i,2:].astype(np.float64).round(5).tolist() }) # 保留5位小数
```

```python
from bson.code import Code
# mapreduce 查询
def mapreduce(freq_input, phi_input):
    # map 函数
    mapfun = Code("""
                    function(){
                        emit({freqs:this.freq, phis:this.phi}, {values: this.vals});
                    }
                """)

    # reduce 函数
    reducefun = Code("""
                    function(key, value) {
                        var res = { values: value };
                        return res;
                    }
                """)
    result = my_set.map_reduce(mapfun, reducefun, query={"freq":freq_input, "phi":phi_input},out="test")
    return result
```

```python
    
# 查询参数
freq_input = 1000000000
phi_input = 0

# mapreduce 查询
result = mapreduce(freq_input, phi_input)
result_num = result.count()
res1 = [doc['value']['values'] for doc in result.find()][0]
result = mapreduce(1200000000, 0)
res2 = [doc['value']['values'] for doc in result.find()][0]
```

```python
# 将 theta 转换为度数
theta = data.values[0][2:].astype(np.float64) * 180 / np.pi
# 频率的所有取值
freq = list(set(data.values[1:, 0].astype(np.int64)))
# phi的所有取值
# 将 phi 转换为度数
phi = list(set(data.values[1:, 1].astype(np.float64) * 180 / np.pi))
```

```python
# 画图
from pyecharts import Line
import pyecharts

pyecharts.configure(global_theme='dark') # 配置主题便于字体显示

attr = np.round(theta,3)
line = Line("theta角度vs电场作图")
line.add("Electrial Field1", attr, res1, is_smooth=True,
         xaxis_name = 'Theta(degree)', yaxis_name = 'Electrical field (V/m)',
         xaxis_name_gap = 30, yaxis_name_gap = 40,
         mark_point=['max', 'min'])
line.add("Electrial Field2", attr, res2, is_smooth=True,
         xaxis_name = 'Theta(degree)', yaxis_name = 'Electrical field (V/m)',
         xaxis_name_gap = 30, yaxis_name_gap = 40,
         mark_point=['max', 'min'])
line
```
